```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# create a data frame that is 100 rows and 100 columns and contains categorical data
# create a 100x100 float matrix
df = pd.DataFrame(np.random.randn(100, 100))
df.insert(0, "Category", np.random.choice(["A", "B", "C", "D"], size=100))
```

Creating a heat map of the data

```python
plt.figure(figsize=(10, 8))
plt.imshow(df.apply(lambda x: pd.factorize(x)[0]), cmap="viridis", aspect="auto")
plt.colorbar()
plt.title("Heatmap Data")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.show()
```

Now create a distribution plot for one column

```python
# creating a distribution plot for one column
plt.figure(figsize=(10, 6))
df[1].hist(bins=30, density=True, alpha=0.5, color="g")
df[1].plot(kind="kde", color="k")
```
